```python
import pandas as pd
import os
import shutil
```

```python
df = pd.read_csv("CycPepDB.csv")
df["Date"] = pd.to_datetime(df["Date"])
df["PDB"] = df["PDB"].apply(lambda x: x.lower())
df["Pep_lengths"] = df["Pep_lengths"].apply(lambda x: x.split(";")[-1].split(":")[-1])
total_structs = len(df)


complexes_df = df.dropna(subset=["Prot_chains", "Link", "Resolution"])
complexes_df = complexes_df[((complexes_df["Method"] == "X-ray") | (complexes_df["Method"] == "NMR")) & (24 >= complexes_df["Pep_lengths"].astype("int32"))]

complexes_df
```

```python
complexes_df.to_csv("filtered_complexes.csv", index=0)
```

```python
for name in complexes_df["PDB"].to_list():
    cif = os.path.join("cpdb_all_cifs", f"{name}.cif.gz")
    shutil.copy(cif, "complexes")
```
